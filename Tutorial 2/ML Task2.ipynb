```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3]
t = [1.2,1.9,3.2]
```

```python
def parameters(x,y):
    a = np.sum(x)
    b = np.sum(t)
    c = np.dot(x,x)
    d = np.dot(x,t)
    N = len(x)
    
    w1 = (b*a - d*N)/(a*a - N*c)
    w0 = (d - w1*c)/a
    return [w0,w1]
```

```python
intercept,slope = parameters(x,t)
y_hat = []
for i in range(len(x)) :
    y_hat.append(slope*x[i] + intercept)
plt.scatter(x,t)
plt.plot(x,y_hat)
```
